```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/sample.csv',sep=';')
```

```python
seqs = df['PARCOURS_BENEF_ID']
seqs
```

```python

history_depth = 3

for idx,row in seqs.items():
    seq = row.split('|')
    seq = [int(i) for i in seq]
    
    X = np.array(seq)
    XN = np.insert(X,0,np.zeros(history_depth))
#     print(XN)
#     print(XN.shape[0])
    
    for x in range(XN.shape[0]-history_depth):  
        print("X:",XN[x:x+history_depth],"Y:",XN[x+history_depth])
```
